```python
# changing the data types
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.head(10)
```

```python

# Function to extract year
def extract_year(release_date):
    split_data = release_date.split(',')
    year_corrected = split_data[-1].strip()[:4]  # Extracting the year part and removing any leading/trailing spaces
    return year_corrected

# Apply the function to the 'released' column
df['year_released'] = df['released'].apply(extract_year)


```
