```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'data' is your DataFrame containing 'Genre' and 'Score' columns

# Get unique genres and count
genres = data['Genre'].unique()
num_genres = len(genres)

# Create a colormap with enough distinct colors for all genres
colors = plt.cm.get_cmap('tab10', num_genres)  # 'tab10' colormap provides 10 distinct colors
legend_colors = [colors(i) for i in range(num_genres)]

# Group the data by 'Genre' and extract scores
grouped_data = [data[data['Genre'] == genre]['Score'] for genre in genres]

# Create the boxplot with specified colors
plt.figure(figsize=(8, 6))  # Set figure size
bp = plt.boxplot(grouped_data, labels=genres, patch_artist=True)

# Set colors for each box and create a legend
legend_boxes = [plt.Rectangle((1,1), 1, 1, color=legend_colors[i]) for i in range(num_genres)]
plt.legend(legend_boxes, genres,loc='upper right',)

for i, box in enumerate(bp['boxes']):
    box.set(facecolor=legend_colors[i])

plt.xlabel('Genre')
plt.ylabel('Score')
plt.title('Distribution of Scores by Genre')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Assuming 'data' is your DataFrame containing the 'Hours_Played' column

plt.figure(figsize=(16, 8))

# Create the histogram
plt.hist(data['Hours_Played'], bins='auto', density=True)

# Calculate mean and standard deviation
mean_hours = data['Hours_Played'].mean()
std_hours = data['Hours_Played'].std()

# Generate normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_hours, std_hours)# y axis values
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('Hours Played')
plt.ylabel('Frequency')
plt.title('Histogram with Normal Distribution')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Assuming 'data' is your DataFrame containing the 'Hours_Played' column

plt.figure(figsize=(16, 8))

# Create the histogram
plt.hist(data['Hours_Played'], bins='auto', density=True, alpha=1, label='Hours Played')

# Calculate mean and standard deviation
mean_hours = data['Hours_Played'].mean()
std_hours = data['Hours_Played'].std()

# Generate normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_hours, std_hours)

# Create a twin Axes
ax2 = plt.gca().twinx()

# Plot the normal distribution curve on the twin Axes
ax2.plot(x, p, 'r-', label='Normal Distribution')

# Set labels and legend
ax2.set_ylabel('Normal Distribution', color='r')
plt.xlabel('Hours Played')
plt.ylabel('Frequency')
plt.title('Histogram with Normal Distribution')

# Show legend for both plots
plt.legend(loc='upper right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import itertools  # Import itertools for cycling colors

# Assuming you have loaded your data into a Pandas DataFrame named 'data'

# Cleaning the 'Score' column
data['Score'] = pd.to_numeric(data['Score'], errors='coerce')  # Convert to numeric, coerce invalid values to NaN

# Removing rows with NaN values in the 'Score' column
data = data.dropna(subset=['Score'])

# Rest of the code for plotting
unique_platforms = data['Platform'].unique()
colors = itertools.cycle(['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink'])  # Cycle through colors

plt.figure(figsize=(10, 6))

for platform in unique_platforms:
    platform_data = data[data['Platform'] == platform]
    plt.scatter(
        platform_data['Score'],
        platform_data['Hours_Played'],
        s=platform_data['Hours_Played'],
        alpha=platform_data['Score'] / data['Score'].max(),
        label=platform,
        c=next(colors)  # Get the next color from the cycle
    )

plt.xlabel('Score')
plt.ylabel('Hours Played')
plt.title('Scatter Plot of Hours Played vs Score by Platform')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew

# Assuming 'cars_df' is your DataFrame with a 'Price' column
# Replace this with your actual dataset

# Create a histogram of prices
plt.figure(figsize=(8, 6))
plt.hist(cars_df['Price'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Vehicle Prices')

# Calculate skewness
skewness = skew(cars_df['Price'])

# Calculate mean and median
mean_price = np.mean(cars_df['Price'])
median_price = np.median(cars_df['Price'])

# Plot mean and median lines
plt.axvline(mean_price, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_price:.2f}')
plt.axvline(median_price, color='green', linestyle='dashed', linewidth=1, label=f'Median: {median_price:.2f}')

# Show skewness direction
if skewness > 0:
    skew_direction = 'Positive Skewness'
elif skewness < 0:
    skew_direction = 'Negative Skewness'
else:
    skew_direction = 'No Skewness'
    
plt.text(0.95, 0.85, skew_direction, fontsize=10, ha='right', va='top', transform=plt.gca().transAxes)

# Show legend
plt.legend()

plt.grid(True)
plt.tight_layout()

# Show the histogram with mean, median lines, and skewness direction annotation
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generating a dataset with negative skewness
np.random.seed(42)
data_negative_skew = np.random.exponential(scale=10, size=1000) * -1

# Calculate mean and median
mean_negative_skew = np.mean(data_negative_skew)
median_negative_skew = np.median(data_negative_skew)

# Plotting histogram and KDE plot
plt.figure(figsize=(8, 6))
sns.histplot(data_negative_skew, kde=True, color='skyblue')
plt.axvline(mean_negative_skew, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_negative_skew:.2f}')
plt.axvline(median_negative_skew, color='green', linestyle='dashed', linewidth=1, label=f'Median: {median_negative_skew:.2f}')
plt.legend()
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram with KDE for Negative Skewness')

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generating a dataset with negative skewness using chi-squared distribution
np.random.seed(42)
data_negative_skew_pos = np.random.chisquare(df=3, size=1000)

# Apply transformation to induce negative skewness
data_negative_skew_pos = 10 - np.sqrt(data_negative_skew_pos)

# Calculate mean and median
mean_negative_skew_pos = np.mean(data_negative_skew_pos)
median_negative_skew_pos = np.median(data_negative_skew_pos)

# Plotting histogram and KDE plot
plt.figure(figsize=(8, 6))
sns.histplot(data_negative_skew_pos, kde=True, color='skyblue')
plt.axvline(mean_negative_skew_pos, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_negative_skew_pos:.2f}')
plt.axvline(median_negative_skew_pos, color='green', linestyle='dashed', linewidth=1, label=f'Median: {median_negative_skew_pos:.2f}')
plt.legend()
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram with KDE for Negative Skewness (Positive Values)')

# Show the plot
plt.show()
```
